```python
from fears.population import Population
import numpy as np

np.random.seed(2022) # set random seed for reproducibility

options = {'max_dose':1, # max drug concentration
           'n_sims':1,
           'k_abs':0.005, # rate of change of drug concentration
           'k_elim':0,
           'passage':True, # whether to simulate cell passaging (periodic dilution of cells)
           'dwell':True, # add a delay before the addition of drug (for the accumulation of mutations)
           'dwell_time':24*3, 
           'carrying_cap':10**9}

p = Population(fitness_data='from_file',**options)
p.simulate();
```

```python
p.print_params()
```
